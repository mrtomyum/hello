```python
for key in score.keys():
  print(key)
```

```python
for key in score.keys():
  print(f'{key}: {score[key]}')
```

```python
del score['B']
score
```

```python
score = {
    'class 1/1': {'A1': 17, 'B1': 15, 'C1': 19},
    'class 1/2': {'A2': 16, 'B2': 17, 'C2': 20}}
score
```

```python
score['class 1/1']
```

```python
score['class 1/1']['A1']
```

```python
for class_name in score.keys():
  for student_name in score[class_name].keys():
    print(f'{class_name}: {student_name}: {score[class_name][student_name]}')
```
